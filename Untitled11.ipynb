```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import streamlit as st
from wordcloud import WordCloud
from datetime import datetime
import re
from collections import Counter
import warnings
warnings.filterwarnings('ignore')

# Set plotting style
plt.style.use('ggplot')
sns.set_palette("viridis")
```

```python
# Load the metadata.csv file
try:
    df = pd.read_csv('metadata.csv')
    print("‚úÖ Successfully loaded metadata.csv")
    print(f"üìä Dataset contains {df.shape[0]} rows and {df.shape[1]} columns")
except FileNotFoundError:
    print("‚ùå metadata.csv not found. Please make sure the file is in the current directory.")
    print("üí° You can download it from: https://www.kaggle.com/allen-institute-for-ai/CORD-19-research-challenge")
    # Create sample data for demonstration purposes
    print("üìã Creating sample data for demonstration...")
    
    # Generate sample data
    titles = [
        "COVID-19 transmission dynamics and control measures",
        "Clinical characteristics of coronavirus patients",
        "Vaccine development for SARS-CoV-2",
        "Impact of pandemic on mental health",
        "Economic consequences of COVID-19 lockdowns",
        "Diagnostic testing methods for COVID-19",
        "Treatment options for severe COVID-19 cases",
        "Genomic analysis of SARS-CoV-2 variants",
        "Public health response to COVID-19 pandemic",
        "Modeling the spread of infectious diseases"
    ]
    
    journals = ["The Lancet", "Nature", "Science", "JAMA", "NEJM", "BMJ", "PLOS One", "Cell", "PNAS", "Elsevier"]
    
    sample_data = []
    for i in range(1000):
        publish_date = pd.to_datetime('2020-01-01') + pd.Timedelta(days=np.random.randint(0, 365*3))
        sample_data.append({
            'title': np.random.choice(titles),
            'journal': np.random.choice(journals + [np.nan], p=[0.9, 0.1]),
            'publish_time': publish_date.strftime('%Y-%m-%d'),
            'abstract': f"Abstract text for paper {i} about COVID-19 research" if np.random.random() > 0.2 else np.nan,
            'authors': f"Author {i}A, Author {i}B, Author {i}C" if np.random.random() > 0.1 else np.nan,
            'url': f"https://example.com/paper/{i}" if np.random.random() > 0.3 else np.nan
        })
    
    df = pd.DataFrame(sample_data)
    print("‚úÖ Sample data created for demonstration")

# Display basic information about the dataset
print("üìã First few rows of the dataset:")
df.head()
```

```python
# Check DataFrame dimensions and structure
print(f"üìä Dataset shape: {df.shape}")
print(f"üî§ Columns: {list(df.columns)}")
print("\nüìù Data types:")
print(df.dtypes)
```

```python
# Check for missing values
print("‚ùì Missing values in each column:")
missing_data = df.isnull().sum()
missing_percentage = (df.isnull().sum() / len(df) * 100)
missing_info = pd.DataFrame({
    'Missing Values': missing_data,
    'Percentage': missing_percentage
}).sort_values('Missing Values', ascending=False)

missing_info
```

```python
# Generate basic statistics for numerical columns (if any)
print("üìà Basic statistics:")
df.describe(include='all')
```

```python
# Create a cleaned version of the dataset
df_clean = df.copy()

# Handle missing values
print("üßπ Handling missing values...")

# For title - we can't fill missing titles, so we'll drop rows with missing titles
initial_count = len(df_clean)
df_clean = df_clean.dropna(subset=['title'])
print(f"  - Dropped {initial_count - len(df_clean)} rows with missing titles")

# For abstract - fill with empty string
df_clean['abstract'] = df_clean['abstract'].fillna('')

# For journal - fill with 'Unknown'
df_clean['journal'] = df_clean['journal'].fillna('Unknown')

# For publish_time - convert to datetime and extract year
print("üìÖ Converting publish_time to datetime...")
try:
    df_clean['publish_time'] = pd.to_datetime(df_clean['publish_time'], errors='coerce')
    # Drop rows with invalid dates
    date_initial_count = len(df_clean)
    df_clean = df_clean.dropna(subset=['publish_time'])
    print(f"  - Dropped {date_initial_count - len(df_clean)} rows with invalid dates")
    
    # Extract year for time-based analysis
    df_clean['year'] = df_clean['publish_time'].dt.year
    print(f"  - Extracted year from publish_time")
except Exception as e:
    print(f"  - Error processing dates: {e}")
    # Create a year column if date processing fails
    df_clean['year'] = 2020  # Default year

# Create abstract word count
df_clean['abstract_word_count'] = df_clean['abstract'].apply(lambda x: len(str(x).split()))
print(f"  - Created abstract_word_count column")

print(f"‚úÖ Cleaned dataset shape: {df_clean.shape}")
```

```python
# Count papers by publication year
print("üìä Papers by publication year:")
year_counts = df_clean['year'].value_counts().sort_index()
print(year_counts)

# Plot publications over time
plt.figure(figsize=(10, 6))
year_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Publications by Year')
plt.xlabel('Year')
plt.ylabel('Number of Publications')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Identify top journals publishing COVID-19 research
print("üìö Top journals publishing COVID-19 research:")
top_journals = df_clean['journal'].value_counts().head(10)
print(top_journals)

# Create a bar chart of top publishing journals
plt.figure(figsize=(12, 6))
top_journals.plot(kind='bar', color='lightcoral')
plt.title('Top 10 Journals by Number of COVID-19 Publications')
plt.xlabel('Journal')
plt.ylabel('Number of Publications')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Find most frequent words in titles
print("üî§ Analyzing most frequent words in titles...")

# Function to clean text and extract words
def clean_text(text):
    # Convert to lowercase and remove punctuation
    text = re.sub(r'[^\w\s]', '', str(text).lower())
    # Remove stopwords (simple approach)
    stopwords = {'the', 'and', 'of', 'in', 'to', 'a', 'for', 'on', 'with', 'by', 'an', 'at'}
    words = [word for word in text.split() if word not in stopwords and len(word) > 2]
    return words

# Extract all words from titles
all_words = []
for title in df_clean['title']:
    all_words.extend(clean_text(title))

# Count word frequencies
word_freq = Counter(all_words)
common_words = word_freq.most_common(20)

print("üìä Most common words in titles:")
for word, count in common_words:
    print(f"  {word}: {count}")

# Create a bar chart of common words
plt.figure(figsize=(12, 6))
words, counts = zip(*common_words)
plt.bar(words, counts, color='lightgreen')
plt.title('Top 20 Most Frequent Words in Paper Titles')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Generate a word cloud of paper titles
print("‚òÅÔ∏è Generating word cloud...")

# Combine all titles into a single string
all_titles = ' '.join(df_clean['title'].astype(str))

# Create and generate a word cloud image
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_titles)

# Display the generated image
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Paper Titles')
plt.tight_layout()
plt.show()
```

```python
# Plot distribution of paper counts by source (journal)
plt.figure(figsize=(10, 6))
journal_counts = df_clean['journal'].value_counts()
plt.hist(journal_counts, bins=20, color='purple', alpha=0.7)
plt.title('Distribution of Paper Counts by Journal')
plt.xlabel('Number of Papers')
plt.ylabel('Number of Journals')
plt.axvline(journal_counts.mean(), color='red', linestyle='dashed', linewidth=1, label=f'Mean: {journal_counts.mean():.2f}')
plt.legend()
plt.tight_layout()
plt.show()

print(f"üìà Journal statistics:")
print(f"  - Total journals: {len(journal_counts)}")
print(f"  - Mean papers per journal: {journal_counts.mean():.2f}")
print(f"  - Median papers per journal: {journal_counts.median()}")
print(f"  - Max papers in a journal: {journal_counts.max()}")
```
